```python

 import pandas as pd
import seaborn as sns
import matplotlib as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')
file_path = ('/content/drive/MyDrive/Libro1.xlsx')
data=pd.read_excel(file_path)
```

```python
# prompt: Mediante el marco de datos data: para el correcto analisis descriptivo y graficos

# Check for missing values
print(data.isnull().sum())

# Summary statistics for numerical columns
print(data.describe())

# Value counts for categorical columns
for col in data.select_dtypes(include=['category']):
    print(f"\nValue counts for {col}:")
    print(data[col].value_counts())

# Explore relationships between variables
# Example: correlation matrix for numerical columns
correlation_matrix = data.select_dtypes(include=['number']).corr()
print(correlation_matrix)

# You can further explore relationships with visualizations:
# Example: using matplotlib or seaborn
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='id', y='sub_id', data=data)
plt.title('Scatter plot of id vs sub_id')
plt.show()

# Histogram
plt.figure(figsize=(10, 6))
sns.histplot(data['motivo_reposicion_id'], bins=20)
plt.title('Histogram of motivo_reposicion_id')
plt.show()

# Box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='nombre_categoria', y='id', data=data)
plt.title('Box plot of id by nombre_categoria')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,10))
sns.countplot(x='categoria_id',data=data)

# Corrected the typo 'enumarate' to 'enumerate'
for i,valor in enumerate(data['categoria_id'].value_counts()):
  plt.text(i,valor,valor)

plt.title('Distribución Criterio de Reposicion')
plt.xlabel('Motivo de Reposición')
plt.ylabel('Frecuencia')
plt.show()
```

```python
plt.figure(figsize=(10,10))
ax = sns.countplot(x='motivo_reposicion_id', data=data, palette="viridis", order=data['motivo_reposicion_id'].value_counts().index)

for p in ax.patches:
  ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.title('Distribución de Categorías de Reposición')
plt.xlabel('Categoría')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,8))
data['motivo_reposicion_id'].value_counts().plot.pie(autopct='%1.1f%%', colors=sns.color_palette("viridis", n_colors=len(data['motivo_reposicion_id'].unique())))
plt.title('Proporción de Categorías de Reposición')
plt.ylabel('')  # Quitar etiqueta del eje Y
plt.show()
```

```python
plt.figure(figsize=(12,6))
motivos_frecuentes = data['descripcion'].value_counts().nlargest(10)
sns.barplot(x=motivos_frecuentes.index, y=motivos_frecuentes.values, palette="Blues_d")
plt.title('Top 10 Motivos de Reposición Más Frecuentes')
plt.xlabel('Motivo')
plt.ylabel('Frecuencia')
plt.xticks(rotation=90)
plt.subplots_adjust(bottom=0.2)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='motivo_reposicion_id', y='adjunto', data=data, palette="Set2")
plt.title('Distribución de Adjuntos por Categoría')
plt.xlabel('Categoría')
plt.ylabel('Cantidad de Adjuntos')
plt.xticks(rotation=45)
plt.show()
```

```python
data[data['nombre_categoria'] == 'Rechazos'].to_csv('rechazos.csv', index=False)
data[data['nombre_categoria'] == 'Otros'].to_csv('otros.csv', index=False)
data[data['nombre_categoria'] == 'Reemplazos'].to_csv('reemplazos.csv', index=False)
```

```python
# Get a list of unique values from the 'motivo_reposicion_id' column
unique_motivos = data['motivo_reposicion_id'].unique()

# Iterate over the unique values and create an Excel file for each one
for motivo in unique_motivos:
  data[data['motivo_reposicion_id'] == motivo].to_excel(f'{motivo}.xlsx', index=False)

# Download the generated Excel files
from google.colab import files
for motivo in unique_motivos:
  files.download(f'{motivo}.xlsx')
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='postulacion_id', y='categoria_id', hue='motivo_reposicion_id', data=data, palette="coolwarm")
plt.title('Relación entre Postulaciones y Categorías')
plt.xlabel('Postulación ID')
plt.ylabel('Categoría ID')
plt.legend(title='Categoría', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(8,6))
correlation_matrix = data[['id', 'sub_id', 'postulacion_id', 'usuario_run', 'categoria_id', 'adjunto']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Matriz de Correlación entre Variables Numéricas')
plt.show()
```

```python
plt.figure(figsize=(10,6))
adjuntos_por_categoria = data.groupby('motivo_reposicion_id')['adjunto'].sum()
ax = sns.barplot(x=adjuntos_por_categoria.index, y=adjuntos_por_categoria.values, palette="Spectral")
plt.title('Comparación de Adjuntos por Categoría')
plt.xlabel('Categoría')
plt.ylabel('Total de Adjuntos')
plt.xticks(rotation=45)

for i, v in enumerate(adjuntos_por_categoria):
    ax.text(i, v, str(v), ha='center', va='bottom')

plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.regplot(x='usuario_run', y='adjunto', data=data, scatter_kws={'s':10}, line_kws={"color":"r","alpha":0.7,"lw":2})
plt.title('Relación entre Usuario y Cantidad de Adjuntos')
plt.xlabel('Usuario Run')
plt.ylabel('Adjuntos')
plt.show()
```

```python
plt.figure(figsize=(100,6))
sns.countplot(x='usuario_run', data=data, palette="viridis")
plt.title('cantidad de motivos por run ')
plt.xlabel('Categoría')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45)
plt.show()
```
