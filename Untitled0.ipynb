```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,10))
sns.countplot(x='categoria_id',data=data)

# Corrected the typo 'enumarate' to 'enumerate'
for i,valor in enumerate(data['categoria_id'].value_counts()):
  plt.text(i,valor,valor)

plt.title('Distribución Criterio de Reposicion')
plt.xlabel('Motivo de Reposición')
plt.ylabel('Frecuencia')
plt.show()
```

```python
plt.figure(figsize=(10,10))
ax = sns.countplot(x='motivo_reposicion_id', data=data, palette="viridis", order=data['motivo_reposicion_id'].value_counts().index)

for p in ax.patches:
  ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.title('Distribución de Categorías de Reposición')
plt.xlabel('Categoría')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,8))
data['motivo_reposicion_id'].value_counts().plot.pie(autopct='%1.1f%%', colors=sns.color_palette("viridis", n_colors=len(data['motivo_reposicion_id'].unique())))
plt.title('Proporción de Categorías de Reposición')
plt.ylabel('')  # Quitar etiqueta del eje Y
plt.show()
```

```python
plt.figure(figsize=(12,6))
motivos_frecuentes = data['descripcion'].value_counts().nlargest(10)
sns.barplot(x=motivos_frecuentes.index, y=motivos_frecuentes.values, palette="Blues_d")
plt.title('Top 10 Motivos de Reposición Más Frecuentes')
plt.xlabel('Motivo')
plt.ylabel('Frecuencia')
plt.xticks(rotation=90)
plt.subplots_adjust(bottom=0.2)
plt.show()
```

```python
# Get a list of unique values from the 'motivo_reposicion_id' column
unique_motivos = data['motivo_reposicion_id'].unique()

# Iterate over the unique values and create an Excel file for each one
for motivo in unique_motivos:
  data[data['motivo_reposicion_id'] == motivo].to_excel(f'{motivo}.xlsx', index=False)

# Download the generated Excel files
from google.colab import files
for motivo in unique_motivos:
  files.download(f'{motivo}.xlsx')
```

```python
plt.figure(figsize=(8,6))
correlation_matrix = data[['id', 'sub_id', 'postulacion_id', 'usuario_run', 'categoria_id', 'adjunto']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Matriz de Correlación entre Variables Numéricas')
plt.show()
```

```python
plt.figure(figsize=(10,6))
adjuntos_por_categoria = data.groupby('motivo_reposicion_id')['adjunto'].sum()
ax = sns.barplot(x=adjuntos_por_categoria.index, y=adjuntos_por_categoria.values, palette="Spectral")
plt.title('Comparación de Adjuntos por Categoría')
plt.xlabel('Categoría')
plt.ylabel('Total de Adjuntos')
plt.xticks(rotation=45)

for i, v in enumerate(adjuntos_por_categoria):
    ax.text(i, v, str(v), ha='center', va='bottom')

plt.show()
```

```python
# prompt: creame un celda para responder de fomra automatica de acuerdo a los criterios

# @title Respuesta Automática
import pandas as pd

# Función para generar la respuesta automática
def generar_respuesta(row):
  """
  Genera una respuesta automática basada en los criterios del DataFrame.

  Args:
    row: Una fila del DataFrame.

  Returns:
    Una cadena de texto con la respuesta automática.
  """
  motivos = row['motivos_reposicion_id']
  descripcion = row['descripcion']
  criterio = row['criterio']
  texto_motivo = row['texto_motivo']

  # Lógica para generar la respuesta
  respuesta = "Estimado/a usuario/a,\n\n"
  respuesta += "Hemos recibido su solicitud de reposición.\n\n"
  respuesta += "De acuerdo con la información proporcionada, hemos identificado los siguientes motivos:\n"
  for i in range(len(motivos)):
    respuesta += f"- {motivos[i]} - {descripcion[i]} ({criterio[i]})\n"
  respuesta += "\n"
  respuesta += "En relación a su solicitud, le informamos que:\n"
  # Aquí puedes agregar la lógica para generar la respuesta basada en los motivos, descripción y criterio.
  # Por ejemplo, puedes usar condicionales para determinar la respuesta adecuada.
  # Si no hay información específica para generar la respuesta, puedes agregar una respuesta genérica.
  respuesta += "Gracias por su consulta.\n\n"
  respuesta += "Atentamente,\n"
  respuesta += "El equipo de atención al cliente."

  return respuesta

# Agrupa los datos por 'usuario_run' y crea una lista de 'motivo_reposicion_id', 'descripcion' y 'criterio_social' para cada usuario
motivos_por_usuario = data.groupby('usuario_run').agg(
    motivos_reposicion_id=('motivo_reposicion_id', list),
    descripcion=('descripcion', list),
    criterio=('nombre_categoria', list),
    texto_motivo=('texto_motivo', list)
).reset_index()

# Aplica la función generar_respuesta a cada fila del DataFrame
motivos_por_usuario['respuesta_automatica'] = motivos_por_usuario.apply(generar_respuesta, axis=1)

# Muestra el DataFrame resultante con la columna de respuesta automática
print(motivos_por_usuario)
```
